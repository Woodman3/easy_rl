```python
import numpy as np
import gymnasium as gym
```

```python
class Config:
    def __init__(self):
        self.gamma = 0.9
        self.alpha = 0.1
        self.epsilon = 0.1
        self.num_episodes = 1000
        self.max_steps = 1000
        self.num_states = 48
        self.num_actions = 4
        self.learning_rate = 0.1
        self.train_eps = 100

env = gym.make("CliffWalking-v0")  # 0 up, 1 right, 2 down, 3 left
cfg = Config()
```

```python
from agent import Q_greedy
agent = Q_greedy(env.observation_space.n, env.action_space.n, cfg.learning_rate,cfg.gamma)
```

```python
for i_ep in range(cfg.train_eps): # train_eps: 训练的最大episodes数
    ep_reward = 0  # 记录每个episode的reward
    states =[]
    actions =[]
    rewards =[]
    state,_ = env.reset()  # 重置环境, 重新开一局（即开始新的一个episode）

    while True:
        action = agent.choose_action(state)  # 根据算法选择一个动作
        next_state, reward,terminated, truncated, info = env.step(action)  # 与环境进行一次动作交互
        states.append(state)
        actions.append(action)
        rewards.append(reward)
        state = next_state  # 存储上一个观察值
        print(f"({state},{state/12},{state%12})")
        ep_reward += reward
        if terminated:
            break
    agent.update(states, actions, rewards)  # Q-learning算法更新
    rewards.append(ep_reward)
    if ma_rewards:
        ma_rewards.append(ma_rewards[-1]*0.9+ep_reward*0.1)
    else:
        ma_rewards.append(ep_reward)
    print("Episode:{}/{}: reward:{:.1f}".format(i_ep+1, cfg.train_eps,ep_reward))
```
